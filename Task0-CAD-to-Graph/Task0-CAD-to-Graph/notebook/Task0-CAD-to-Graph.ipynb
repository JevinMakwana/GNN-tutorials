```python
import pyvista as pv
pv.set_jupyter_backend('html')

# Camshaft_Cap
stl_obj_path = "/content/drive/MyDrive/TVS/Task0-CAD-to-Graph/Camshaft_Cap.stl"
# stl_obj_path = "/content/drive/MyDrive/TVS/Task0-CAD-to-Graph/Blade_Type_5.stl"
mesh = pv.read(stl_obj_path)
mesh.plot()
```

```python
# The .stl object used here contains surface data only.
# But if object contains volume data and we want to extract surface data then
# following line of code is used:-
surface_mesh = mesh.extract_surface(algorithm='dataset_surface')
```

```python
print("Number of points in given object:", mesh.n_points)
print("Number of cells in given object:", mesh.n_cells)
```

```python
mesh_points = mesh.points
print("mesh_points: ",mesh_points)
```

```python
mesh_faces = mesh.faces
print("mesh faces in compact '3, p1v1, p1v2, p1v3, ...' form:",mesh_faces)

mesh_faces = mesh_faces.reshape(-1,4)[:, 1:]
print("mesh faces in original '(v1, v2, v3)' form:",mesh_faces)
```

#Create edges

```python
# THIS CODE MIGHT ADD REDUNDANT EDGES. i.e: 1-2 AND 2-1
# edges = []
# for tri in mesh_faces:
#   a, b, c = tri
#   edges.extend([
#       [a, b],
#       [b, c],
#       [c, a]
#   ])

# print(type(edges))
# print(edges)
```

```python
# BY USING SORTED TUPLE, REDUNDANCY IS AVOIDED
edges = set()

for tri in mesh_faces:
    a, b, c = tri

    edges.update([
        tuple(sorted((a, b))),
        tuple(sorted((b, c))),
        tuple(sorted((c, a)))
    ])

print(edges)
```

#Create adjacency matrix

```python
import numpy as np
num_nodes = mesh.n_points

adj_matrix = np.zeros((num_nodes, num_nodes), dtype=int)

for u, v in edges:
    adj_matrix[u][v] = 1
    adj_matrix[v][u] = 1

print(adj_matrix)
```

#Create edge_index


```python
edge_index = np.array(list(edges)).T
# [0] => source nodes, [1] => destinaiton nodes
print(edge_index) 
print(edge_index.shape)
```

#Extract norms data. Norm => direction of surface

```python
mesh_with_normals = mesh.compute_normals(
    cell_normals=False,
    point_normals=True
)

normals = mesh_with_normals.point_data['Normals']

print(normals)
print(normals.shape)
```

#Finally, create features-list [[x_coord, y_coord, z_coord, x_norm, y_norm, z_norm]]

```python
coords = mesh.points
normals = mesh_with_normals.point_data['Normals']

node_features = np.hstack([coords, normals])

print(node_features)
print(node_features.shape)
```

#Compare graph with mesh-structure

```python
# convert edges into PyVista line format
lines = []

for u, v in edges:
    lines.extend([2, u, v])

lines = np.array(lines)

# create graph mesh
graph = pv.PolyData()
graph.points = mesh_points
graph.lines = lines

# interactive plot
plotter = pv.Plotter(notebook=True)

# actual mesh
plotter.add_mesh(
    mesh,
    color='lightblue',
    opacity=0.4,
    show_edges=False,
    label = "original mesh"
)

# graph edges
plotter.add_mesh(
    graph,
    color='red',
    line_width=1,
    label = "graph edges"
)

# graph nodes
plotter.add_points(
    mesh_points,
    color='black',
    point_size=5,
    label = "graph points/nodes/vertices",
    render_points_as_spheres=True
)

plotter.add_legend()


# plotter.show(jupyter_backend='html')
```

```python
from torch_geometric.data import Data
import torch

edge_index = torch.tensor(list(edges), dtype=torch.long).t().contiguous()

data = Data(
    x=torch.tensor(node_features, dtype=torch.float),
    edge_index=edge_index
)

print("data: ", data)
```

```python
# extract points from pyG-GRAPH
pyG_points = data.x[:, :3].numpy()
print("pyG_points.shape: ", pyG_points.shape)

# extract candidate-edges-idx from pyG-GRAPH
pyG_edge_idx = data.edge_index.numpy() # sourse-destination format yet
print("pyG_edge_idx", pyG_edge_idx.shape)
```

```python
# create actual edges from sourse-destinaiton edge-list
pyG_edges = []

for idx in range(len(pyG_edge_idx[0])):
  v = pyG_edge_idx[0][idx]
  u = pyG_edge_idx[1][idx]
  pyG_edges.extend([2, v, u])

pyG_edges = np.array(pyG_edges)
pyG_edges[:5]
# 2, n1, n2, 2, n3, n4, ...
# 3,n1,n2,n3, 3,n4,n5,n6, ...
```

```python
recreated_pvGraph = pv.PolyData()
recreated_pvGraph.points = pyG_points
recreated_pvGraph.lines = pyG_edges

plotter = pv.Plotter(notebook=True)

plotter.add_mesh(
    mesh,
    color='lightblue',
    opacity=0.35,
    label='Original Mesh'
)

plotter.add_mesh(
    recreated_pvGraph,
    color='red',
    line_width=1,
    label='Graph Edges'
)

plotter.add_points(
    recreated_pvGraph.points,
    color='black',
    line_width=1,
    label='Graph Points'
)

plotter.add_axes()

plotter.show(jupyter_backend='html')
```

#Create .vtm file

```python
# ALL CODE TOGATHER

import pyvista as pv
import numpy as np

# 1. LOAD STL FILE
pv.set_jupyter_backend('html')
stl_obj_path = "/content/drive/MyDrive/TVS/Task0-CAD-to-Graph/Blade_Type_5.stl"
mesh = pv.read(stl_obj_path)

print("Number of points:", mesh.n_points)
print("Number of cells:", mesh.n_cells)
# ---------------------------------------------------


# 2. EXTRACT POINTS (VERTICES)
points = mesh.points
print("Points shape:", points.shape)
# ---------------------------------------------------


# 3. EXTRACT FACES
faces = mesh.faces

# Convert compact form:
# [3, v1, v2, v3, 3, v1, v2, v3 ...]
# into:
# [[v1,v2,v3],
#  [v1,v2,v3]]

faces = faces.reshape(-1, 4)[:, 1:]

print("Faces shape:", faces.shape)
# ---------------------------------------------------


# 4. CREATE GRAPH EDGES
edges = set()

for tri in faces:
    a, b, c = tri

    edges.update([
        tuple(sorted((a, b))),
        tuple(sorted((b, c))),
        tuple(sorted((c, a)))
    ])

print("Number of unique edges:", len(edges))
# ---------------------------------------------------


# 5. COMPUTE SURFACE NORMALS
mesh_with_normals = mesh.compute_normals(
    cell_normals=False,
    point_normals=True
)

normals = mesh_with_normals.point_data['Normals']
print("Normals shape:", normals.shape)
# ---------------------------------------------------


# 6. CREATE NODE FEATURES

# Features:
# [x, y, z, nx, ny, nz]

node_features = np.hstack([points, normals])
print("Node feature shape:", node_features.shape)
print("\nSample node feature:")
print(node_features[0])
# ---------------------------------------------------



# 7. CREATE GRAPH POLYDATA

# PyVista line format:
# [2, p1, p2, 2, p1, p2 ...]

lines = []

for u, v in edges:
    lines.extend([2, u, v])

lines = np.array(lines)

graph = pv.PolyData()
graph.points = points
graph.lines = lines
# ---------------------------------------------------



# 8. ADD FEATURES TO MESH
mesh.point_data['Normals'] = normals
mesh.point_data['NodeFeatures'] = node_features
# ---------------------------------------------------



# 9. CREATE MULTIBLOCK DATASET
multi = pv.MultiBlock()
multi['mesh'] = mesh
multi['graph'] = graph
# ---------------------------------------------------


# 10. SAVE AS .VTM
vtm_output_path = "/content/drive/MyDrive/TVS/Task0-CAD-to-Graph/cad_graph.vtm"
multi.save(vtm_output_path)
print(f"\nSaved .vtm file at: {vtm_output_path}")
# ---------------------------------------------------



# 11. INTERACTIVE VISUALIZATION
plotter = pv.Plotter(notebook=True)

# original mesh
plotter.add_mesh(
    mesh,
    color='lightblue',
    opacity=0.4
)

# graph edges
plotter.add_mesh(
    multi['mesh'],
    color='red',
    line_width=1
)

# graph nodes
plotter.add_points(
    points,
    color='black',
    point_size=5,
    render_points_as_spheres=True
)

plotter.show(jupyter_backend='html')
```
